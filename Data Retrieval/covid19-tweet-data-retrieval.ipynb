```
# Define the search term and the date_since date as variables
search_words = "#Covid-19"
date_since = "2021-02-01"

```

```
new_search = search_words + " -filter:retweets"
new_search
# '#wildfires -filter:retweets'
```

```
tweets = tw.Cursor(api.search, 
                           q=new_search,
                           lang="en",
                           tweet_mode ='extended',
                           since=date_since).items(5000)

tweets_locs = [[tweet.created_at,tweet.id,tweet.full_text,tweet.entities['urls'],tweet.user.name] for tweet in tweets]
tweets_locs
```

```
tweet_text = pd.DataFrame(data=tweets_locs, 
                    columns=['created_at', "tweet_id","text","url","username"])
tweet_text
```
